```python
import os
import shutil
import pytesseract
from PIL import Image
```

```python
# Set the Tesseract path (change it according to your Colab environment if necessary)
pytesseract.pytesseract.tesseract_cmd = "/usr/bin/tesseract"
```

```python
# Define paths for input and output folders
input_folder = "/content/drive/MyDrive/Image Classification 2/images"
output_folder = "/content/drive/MyDrive/Image Classification 2/tender_images"
```

```python
# Load the list of words from the dictionary
with open("/content/drive/MyDrive/Image Classification 2/words.txt", "r") as f:
    word_list = [word.strip() for word in f.readlines()]
```

```python
word_list
```

```python
# Function to perform OCR on an image and extract text
def perform_ocr(image_path):
    # Open the image using PIL
    image = Image.open(image_path)

    # Perform OCR using Tesseract
    text = pytesseract.image_to_string(image)

    return text
```

```python
# Iterate over the images in the input folder
for image_name in os.listdir(input_folder):
    image_path = os.path.join(input_folder, image_name)

    # Perform OCR to extract text from the image
    extracted_text = perform_ocr(image_path)

    # Check if any of the words match the extracted text
    found_words = [word for word in word_list if word in extracted_text]

    if found_words:
        # Move the image to the output folder
        output_path = os.path.join(output_folder, image_name)
        shutil.move(image_path, output_path)
    else:
        # Image does not contain the specified words, keep it in the input folder
        pass
```
